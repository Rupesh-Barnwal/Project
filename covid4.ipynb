```python
from tkinter import *
import tkinter
#import tkinter as tk
from tkinter import ttk
import time
import tkinter.messagebox as MessageBox
import mysql.connector as mysql
from functools import partial


#graph = {'H1':{'H5':6,'H6':8,'H8':15},
#         'H2':{'H3':2,'H4':6,'H6':2,'H7':7,'H9':3},
#         'H3':{'H2':2,'H8':3,'H6':9,'H4':8},
#         'H4':{'H2':6,'H3':8,'H8':25,'H7':5},
#         'H5':{'H1':6,'H6':2,'H9':6,'H10':5},
#         'H6':{'H1':8,'H2':2,'H3':9,'H5':2,'H8':20,'H9':4},
#         'H7':{'H2':7,'H4':5},
#         'H8':{'H1':15,'H3':3,'H4':25,'H6':20},
#         'H9':{'H2':3,'H6':4,'H5':6,'H10':7},
#         'H10':{'H5':5,'H9':7}
#        }

graph = {'Jubilee More':{'ESI Hospitals':1,'Kalla More':6,'Kalyanpur Housing':5,'Kulti':18},
         'ESI Hospitals':{'Jubilee More':1,'HLG':2},
         'HLG':{'ESI Hospitals':2,'Shristi Nagar':1,'Kalyanpur Housing':2,'Bhagat Singh More':3},
         'Shristi Nagar':{'HLG':1},
         'Kalyanpur Housing':{'HLG':2,'Jubilee More':5,'Dhadka':4},
         'Dhadka':{'Kalyanpur Housing':4,'Railpaar':2},
         'Railpaar':{'Dhadka':2,'Asansol Railway Station':1},
         'Asansol Railway Station':{'Railpaar':1,'City Bus Stand':2},
         'City Bus Stand':{'Asansol Railway Station':2,'Ashram More':2,'Hutton Road':2,'Chelidanga':3},
         'Chelidanga':{'City Bus Stand':3,'S.B. Gorai Road':3,'BNR More':1},
         'BNR More':{'Chelidanga':1,'Court Area':3,'Bhagat Singh More':1},
         'Bhagat Singh More':{'BNR More':1,'Sterling Hospital':4,'Kumarpur':2,'HLG':3},
         'Kumarpur':{'Bhagat Singh More':2,'Neamatpur':5},
         'Neamatpur':{'Kumarpur':5,'Kulti':4,'New Town':5},
          'Kulti':{'Neamatpur':4,'Barakar':4,'Jubilee More':18},
          'Barakar':{'Kulti':4},
          'New Town':{'Neamatpur':5,'Radha Nagar':1},
          'Radha Nagar':{'New Town':1,'Galaxy Mall':1},
          'Galaxy Mall':{'Radha Nagar':1,'Sterling Hospital':2,'Court Area':2,'Burnpur Railway Station':2},
          'Burnpur Railway Station':{'Galaxy Mall':2,'Burnpur Bus Stand':2},
          'Burnpur Bus Stand':{'Burnpur Railway Station':2,'Nehru Park':4},
          'Nehru Park':{'Burnpur Bus Stand':4},
          'Sterling Hospital':{'Bhagat Singh More':4,'Galaxy Mall':2,'Court Area':1},
          'Court Area':{'Sterling Hospital':1,'Galaxy Mall':2,'BNR More':3,'S.B. Gorai Road':3},
          'S.B. Gorai Road':{'Court Area':3,'Burnpur Bus Stand':3,'Chelidanga':3,'Ismile More':3},
          'Ismile More':{'Hutton Road':2,'S.B. Gorai Road':3,'Asansol District Hospital':2,'Rehmat Nagar':2},
          'Rehmat Nagar':{'Ismile More':2},
          'Hutton Road':{'Ismile More':2,'City Bus Stand':2},
          'Asansol District Hospital':{'Ismile More':2,'Murgasol':3,'Mohishila Colony':3},
          'Mohishila Colony':{'Asansol District Hospital':3,'Murgasol':3},
          'Murgasol':{'Asansol District Hospital':3,'Mohishila Colony':3,'Ushagram':1,'Ashram More':2},
          'Ashram More':{'City Bus Stand':2,'Murgasol':2,'Kalla More':3},
          'Kalla More':{'Ashram More':3,'Kalipahari':6},
          'Ushagram':{'Murgasol':1,'Kalipahari':1},
          'Kalipahari':{'Ushagram':1,'Kalla More':6,'Raniganj':10},
          'Raniganj':{'Kalipahari':10,'Anand Lok Hospital':1},
          'Anand Lok Hospital':{'Raniganj':1,'Durgapur City Centre':45},
          'Durgapur City Centre':{'Anand Lok Hospital':45,'Health World Hospital':2,'I Q Hospital':12,'The Mission':14},
          'Health World Hospital':{'Durgapur City Centre':2},
          'I Q Hospital':{'Durgapur City Centre':12},
          'The Mission':{'Durgapur City Centre':14}
        }

count = 0

def find_all_paths(graph, start, end, path=[]):
        global count
        path = path + [start]
        if start == end:
            count = count + 1
            return [path]
        if start not in graph:
            count = count + 1
            return []
        paths = []
        for node in graph[start]:
            if node not in path:
                newpaths = find_all_paths(graph, node, end, path)
                for newpath in newpaths:
                    paths.append(newpath)
        return paths


    
def dijkstra_DSH(graph,start,goal):
    find_all_paths(graph, start, goal)
    shortest_distance = {} #records the cost to reach to that node...Going to be updated as we move along the graph
    track_predecessor = {} #keep track of the path that has led us to this
    unseenNodes = graph #to iterate through the entire graph
    infinity = 99999 #it can be considered as a very large number
    track_path = [] #going to trace our journey to the source node...optimal route
    
    for node in unseenNodes:
        shortest_distance[node] = infinity
    shortest_distance[start] = 0
    
    while unseenNodes:
        min_distance_node = None
        for node in unseenNodes:
            if min_distance_node is None:
                min_distance_node = node
            elif shortest_distance[node] < shortest_distance[min_distance_node]:
                min_distance_node = node
                
        path_options = graph[min_distance_node].items()
    
        for child_node, weight in path_options:
            if weight + shortest_distance[min_distance_node] < shortest_distance[child_node]:
                shortest_distance[child_node] = weight + shortest_distance[min_distance_node]
                track_predecessor[child_node] = min_distance_node
        
        unseenNodes.pop(min_distance_node)
    
    currentNode = goal
    
    while currentNode != start:
        try:
            track_path.insert(0, currentNode)
            currentNode = track_predecessor[currentNode]
            
        except KeyError:
            print("Path is Unreachable")
            break
            
    track_path.insert(0, start)
    
    #dijkstra(graph,'H2','H8')
    
    if shortest_distance[goal] != infinity:
        Find = Tk()
        Find.title("Result")
        Find.geometry('1600x800+0+50')
        l3=Label(Find, text=" All Possible paths are :", font="times 13 bold")
        l3.grid(row=8,column=1,padx=5,pady=5)
        l1=Label(Find, text="But, The Optimal Path is :", font="times 13 bold")
        l1.grid(row=20,column=1,padx=5,pady=5)
        l2=Label(Find, text="Shortest distance is :", font="times 13 bold")
        l2.grid(row=34,column=1,padx=5,pady=5)
        
        col=3
        r=20
        l = len(track_path)
        for x in range(l):
            if(col==19):
                r=24
                col=3
            elif(col==19 and r==24):
                r=28
            a1=Label(Find, text=track_path[x], fg="steel blue", font="times 13 bold", anchor=W)
            a1.grid(row=r,column=col,padx=5,pady=5)
            if x != (l-1):
                a2=Label(Find, text="->", fg="steel blue", font="times 15 bold", anchor=W)
                a2.grid(row=r,column=col+1,padx=5,pady=5)
            col=col+2
         
        a3=Label(Find, text=shortest_distance[goal], fg="steel blue", font="times 15 bold", anchor=W)
        a3.grid(row=34,column=3,padx=5,pady=5)   
        a4=Label(Find, text=count, fg="steel blue", font="times 15 bold", anchor=W)
        a4.grid(row=8,column=3,padx=5,pady=5)
        print("Shortest Distance is "+ str(shortest_distance[goal]))
        print("Optimal Path is "+ str(track_path))
        btnexit=Button(Find,padx=10,pady=10,fg="black",font=('ariel' ,20,'bold'),width=8, text="Okay", bg="azure3", command=Find.destroy)
        btnexit.place(x=550, y=300)
        Find.mainloop()
        #dijkstra_HI(graph,'H8','H5')
        print("count= " + str(count))
        
        
        

        
        
def showind():
    showind1 = Tk()
    showind1.title("Individual Table Details")
    showind1.geometry('800x800+300+100')
    con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
    cursor = con.cursor()
    cursor.execute("select * from form")
    data = cursor.fetchall()
    
    p1=Label(showind1, text="Name", font="time 15 bold")
    p1.grid(row=1,column=0,padx=5,pady=5)
    p2=Label(showind1, text="Age", font="time 15 bold")
    p2.grid(row=1,column=1,padx=5,pady=5)
    p3=Label(showind1, text="Address", font="time 15 bold")
    p3.grid(row=1,column=2,padx=5,pady=5)
    #p4=Label(showind1, text="Blood Group", font="time 15 bold")
    #p4.grid(row=1,column=3,padx=5,pady=5)
    p5=Label(showind1, text="Gender", font="time 15 bold")
    p5.grid(row=1,column=3,padx=5,pady=5)
    p6=Label(showind1, text="Contact", font="time 15 bold")
    p6.grid(row=1,column=4,padx=5,pady=5)
    p7=Label(showind1, text="Email", font="time 15 bold")
    p7.grid(row=1,column=5,padx=5,pady=5)
    p8=Label(showind1, text="Date", font="time 15 bold")
    p8.grid(row=1,column=7,padx=5,pady=5)
    num=2   
    for i in data:
        a1=Label(showind1, text=i[0], fg="steel blue", anchor=W)
        a1.grid(row=num,column=0,padx=5,pady=5)
        a2=Label(showind1, text=i[1], fg="steel blue", anchor=W)
        a2.grid(row=num,column=1,padx=5,pady=5)
        a3=Label(showind1, text=i[2], fg="steel blue", anchor=W)
        a3.grid(row=num,column=2,padx=5,pady=5)
        #a4=Label(showind1, text=i[3], fg="steel blue", anchor=W)
        #a4.grid(row=num,column=3,padx=5,pady=5)
        a5=Label(showind1, text=i[3], fg="steel blue", anchor=W)
        a5.grid(row=num,column=3,padx=5,pady=5)
        a6=Label(showind1, text=i[4], fg="steel blue", anchor=W)
        a6.grid(row=num,column=4,padx=5,pady=5)
        a7=Label(showind1, text=i[5], fg="steel blue", anchor=W)
        a7.grid(row=num,column=5,padx=5,pady=5)
        a8=Label(showind1, text=i[6], fg="steel blue", anchor=W)
        a8.grid(row=num,column=7,padx=5,pady=5)
        num=num+1
    con.close();
    
    
def hosp(): #Manipulate Hospital Data
    def insert():
        id = e_id.get()
        name = e_name.get()
        loc = e_loc.get()
        cont = e_cont.get()
        bg1 = e_bg1.get()
        bg2 = e_bg2.get()
        bg3 = e_bg3.get()
        bg4 = e_bg4.get()
        bg5 = e_bg5.get()
        bg6 = e_bg6.get()
        bg7 = e_bg7.get()
        bg8 = e_bg8.get()
        
        if (id=='' or name=='' or loc=='' or cont=='' or bg1=='' or bg2=='' or bg3=='' or bg4=='' or bg4=='' or bg5=='' or bg6=='' or bg7=='' or bg8==''):
            MessageBox.showinfo("Insert Status", "All fields are required")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("insert into hospital values('"+ id +"','"+ name +"','"+ loc +"','"+ cont +"','"+ bg1 +"','"+ bg2 +"','"+ bg3 +"','"+ bg4 +"','"+ bg5 +"','"+ bg6 +"','"+ bg7 +"','"+ bg8 +"')")
            cursor.execute("commit");
            show()
            MessageBox.showinfo("Insert Status", "Inserted successfully");
        
            e_id.delete(0, 'end')
            e_name.delete(0, 'end')
            e_loc.delete(0, 'end')
            e_cont.delete(0, 'end')
            e_bg1.delete(0, 'end')
            e_bg2.delete(0, 'end')
            e_bg3.delete(0, 'end')
            e_bg4.delete(0, 'end')
            e_bg5.delete(0, 'end')
            e_bg6.delete(0, 'end')
            e_bg7.delete(0, 'end')
            e_bg8.delete(0, 'end')
            con.close();  
            
    def delete():
        if(e_id.get()==''):
            MessageBox.showinfo("Delete Status", "Giving id is compulsory")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("delete from hospital where id='"+ e_id.get() +"'")
            cursor.execute("commit");
            show()
            MessageBox.showinfo("Deleted Status", "Deleted successfully");
            e_id.delete(0, 'end')
            e_name.delete(0, 'end')
            e_loc.delete(0, 'end')
            e_cont.delete(0, 'end')
            e_bg1.delete(0, 'end')
            e_bg2.delete(0, 'end')
            e_bg3.delete(0, 'end')
            e_bg4.delete(0, 'end')
            e_bg5.delete(0, 'end')
            e_bg6.delete(0, 'end')
            e_bg7.delete(0, 'end')
            e_bg8.delete(0, 'end')
            con.close();
            
            
    def get():
        if(e_id.get()==''):
            MessageBox.showinfo("Fetch Status", "Giving id is compulsory")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("select * from hospital where id='"+ e_id.get() +"'")
            data = cursor.fetchall()
        
            for row in data:
                e_name.insert(0, row[1])
                e_loc.insert(0, row[2])
                e_cont.insert(0, row[3])
                e_bg1.insert(0, row[4])
                e_bg2.insert(0, row[5])
                e_bg3.insert(0, row[6])
                e_bg4.insert(0, row[7])
                e_bg5.insert(0, row[8])
                e_bg6.insert(0, row[9])
                e_bg7.insert(0, row[10])
                e_bg8.insert(0, row[11])
            
            con.close();
            
    def update():
        id = e_id.get()
        name = e_name.get()
        loc = e_loc.get()
        cont = e_cont.get()
        bg1 = e_bg1.get()
        bg2 = e_bg2.get()
        bg3 = e_bg3.get()
        bg4 = e_bg4.get()
        bg5 = e_bg5.get()
        bg6 = e_bg6.get()
        bg7 = e_bg7.get()
        bg8 = e_bg8.get()
        
        con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
        cursor = con.cursor()
        cursor.execute("update hospital set name='"+ name +"', loc='"+ loc +"', cont='"+ cont +"', bg1='"+ bg1 +"', bg2='"+ bg2 +"', bg3='"+ bg3 +"', bg4='"+ bg4 +"', bg5='"+ bg5 +"', bg6='"+ bg6 +"', bg7='"+ bg7 +"', bg8='"+ bg8 +"' where id='"+ id +"'")
        cursor.execute("commit");
        show()
        MessageBox.showinfo("Update Status", "Updated successfully");
        e_id.delete(0, 'end')
        e_name.delete(0, 'end')
        e_loc.delete(0, 'end')
        e_cont.delete(0, 'end')
        e_bg1.delete(0, 'end')
        e_bg2.delete(0, 'end')
        e_bg3.delete(0, 'end')
        e_bg4.delete(0, 'end')
        e_bg5.delete(0, 'end')
        e_bg6.delete(0, 'end')
        e_bg7.delete(0, 'end')
        e_bg8.delete(0, 'end')
        con.close();
            
    def show(): #funct for showing the hid and name on the manupulating hsopital data screen
        con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
        cursor = con.cursor()
        cursor.execute("select * from hospital")
        data = cursor.fetchall()
        list.delete(0, list.size())
        
        for row in data:
            insertData = '     ' + row[0]+ '             '+ row[1]
            list.insert(list.size()+1, insertData)
        con.close()
        
        
        
        
        
        
    hosp1 = Tk()
    hosp1.title("Manipulate Hospital Data")
    hosp1.geometry('700x700+450+70')
    h3=tkinter.Label(hosp1,text="Manipulate Hospital Data",font=("Times new roman",20),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h3.pack(side=TOP,fill=X)
    
    id = Label(hosp1,text="Enter id",font=('bold',10))
    id.place(x=20,y=45)
    e_id = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_id.place(x=115,y=45)
    name = Label(hosp1,text="Enter Name",font=('bold',10))
    name.place(x=20,y=75)
    e_name = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_name.place(x=115,y=75)
    loc = Label(hosp1,text="Enter Location",font=('bold',10))
    loc.place(x=20,y=105)
    e_loc = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_loc.place(x=115,y=105)
    cont = Label(hosp1,text="Enter Contact",font=('bold',10))
    cont.place(x=20,y=135)
    e_cont = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_cont.place(x=115,y=135)
    bg1 = Label(hosp1,text="Available A+",font=('bold',10))
    bg1.place(x=20,y=165)
    e_bg1 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg1.place(x=115,y=165)
    bg2 = Label(hosp1,text="             A -",font=('bold',10))
    bg2.place(x=20,y=195)
    e_bg2 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg2.place(x=115,y=195)
    bg3 = Label(hosp1,text="             B+",font=('bold',10))
    bg3.place(x=20,y=225)
    e_bg3 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg3.place(x=115,y=225)
    bg4 = Label(hosp1,text="             B -",font=('bold',10))
    bg4.place(x=20,y=255)
    e_bg4 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg4.place(x=115,y=255)
    bg5 = Label(hosp1,text="             AB+",font=('bold',10))
    bg5.place(x=20,y=285)
    e_bg5 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg5.place(x=115,y=285)
    bg6 = Label(hosp1,text="             AB -",font=('bold',10))
    bg6.place(x=20,y=315)
    e_bg6 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg6.place(x=115,y=315)
    bg7 = Label(hosp1,text="             O -",font=('bold',10))
    bg7.place(x=20,y=345)
    e_bg7 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg7.place(x=115,y=345)
    bg8 = Label(hosp1,text="             O+",font=('bold',10))
    bg8.place(x=20,y=375)
    e_bg8 = Entry(hosp1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_bg8.place(x=115,y=375)
    
    btninsert=Button(hosp1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Insert", bg="azure3", command=insert)
    btninsert.place(x=30, y=430)
    btnupdate=Button(hosp1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Update", bg="azure3", command=update)
    btnupdate.place(x=140, y=430)
    btndelete=Button(hosp1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Delete", bg="azure3", command=delete)
    btndelete.place(x=250, y=430)
    btnget=Button(hosp1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Get", bg="azure3", command=get)
    btnget.place(x=360, y=430)
    btnexit=Button(hosp1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Exit", bg="azure3", command=hosp1.destroy)
    btnexit.place(x=470, y=430)
    
    list = Listbox(hosp1,width=40, height=15)
    list.place(x=400,y=45)
    show()
    
    hosp1.mainloop()
    
def ambu(): #Manipulate Ambulance Data
    def insert():
        id = e_id.get()
        Name = e_name.get()
        Loc = e_loc.get()
        Cont = e_cont.get()
        Small = e_small.get()
        RSVAC = e_rsvac.get()
        RSVNAC = e_rsvnac.get()
        Medium = e_mid.get()
        RMVAC = e_rmvac.get()
        RMVNAC = e_rmvnac.get()
        Large = e_large.get()
        RLVAC = e_rlvac.get()
        RLVNAC = e_rlvnac.get()
        Support = e_support.get()
        RSAC = e_rsac.get()
        RSNAC = e_rsnac.get()
        
        if (id=='' or Name=='' or Loc=='' or Cont=='' or Small=='' or RSVAC=='' or RSVNAC=='' or Medium=='' or RMVAC=='' or RMVNAC=='' or Large=='' or RLVAC=='' or RLVNAC=='' or Support=='' or RSAC=='' or RSNAC==''):
            MessageBox.showinfo("Insert Status", "All fields are required")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("insert into ambulance values('"+ id +"','"+ Name +"','"+ Loc +"','"+ Cont +"','"+ Small +"','"+ RSVAC +"','"+ RSVNAC +"','"+ Medium +"','"+ RMVAC +"','"+ RMVNAC +"','"+ Large +"','"+ RLVAC +"','"+ RLVNAC +"','"+ Support +"','"+ RSAC +"','"+ RSNAC +"')")
            cursor.execute("commit");
            show()
            MessageBox.showinfo("Insert Status", "Inserted successfully");
        
            e_id.delete(0, 'end')
            e_name.delete(0, 'end')
            e_loc.delete(0, 'end')
            e_cont.delete(0, 'end')
            e_small.delete(0, 'end')
            e_rsvac.delete(0, 'end')
            e_rsvnac.delete(0, 'end')
            e_mid.delete(0, 'end')
            e_rmvac.delete(0, 'end')
            e_rmvnac.delete(0, 'end')
            e_large.delete(0, 'end')
            e_rlvac.delete(0, 'end')
            e_rlvnac.delete(0, 'end')
            e_support.delete(0, 'end')
            e_rsac.delete(0, 'end')
            e_rsnac.delete(0, 'end')
            con.close();  
            
    def delete():
        if(e_id.get()==''):
            MessageBox.showinfo("Delete Status", "Giving id is compulsory")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("delete from ambulance where id='"+ e_id.get() +"'")
            cursor.execute("commit");
            show()
            MessageBox.showinfo("Deleted Status", "Deleted successfully");
            e_id.delete(0, 'end')
            e_name.delete(0, 'end')
            e_loc.delete(0, 'end')
            e_cont.delete(0, 'end')
            e_small.delete(0, 'end')
            e_rsvac.delete(0, 'end')
            e_rsvnac.delete(0, 'end')
            e_mid.delete(0, 'end')
            e_rmvac.delete(0, 'end')
            e_rmvnac.delete(0, 'end')
            e_large.delete(0, 'end')
            e_rlvac.delete(0, 'end')
            e_rlvnac.delete(0, 'end')
            e_support.delete(0, 'end')
            e_rsac.delete(0, 'end')
            e_rsnac.delete(0, 'end')
            con.close();
            
            
    def get():
        if(e_id.get()==''):
            MessageBox.showinfo("Fetch Status", "Giving id is compulsory")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("select * from ambulance where id='"+ e_id.get() +"'")
            data = cursor.fetchall()
        
            for row in data:
                e_name.insert(0, row[1])
                e_loc.insert(0, row[2])
                e_cont.insert(0, row[3])
                e_small.insert(0, row[4])
                e_rsvac.insert(0, row[5])
                e_rsvnac.insert(0, row[6])
                e_mid.insert(0, row[7])
                e_rmvac.insert(0, row[8])
                e_rmvnac.insert(0, row[9])
                e_large.insert(0, row[10])
                e_rlvac.insert(0, row[11])
                e_rlvnac.insert(0, row[12])
                e_support.insert(0, row[13])
                e_rsac.insert(0, row[14])
                e_rsnac.insert(0, row[15])
            con.close();
            
    def update():
        id = e_id.get()
        Name = e_name.get()
        Loc = e_loc.get()
        Cont = e_cont.get()
        Small = e_small.get()
        RSVAC = e_rsvac.get()
        RSVNAC = e_rsvnac.get()
        Medium = e_mid.get()
        RMVAC = e_rmvac.get()
        RMVNAC = e_rmvnac.get()
        Large = e_large.get()
        RLVAC = e_rlvac.get()
        RLVNAC = e_rlvnac.get()
        Support = e_support.get()
        RSAC = e_rsac.get()
        RSNAC = e_rsnac.get()
        
        con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
        cursor = con.cursor()
        cursor.execute("update ambulance set Name='"+ name +"', Loc='"+ Loc +"', Cont='"+ Cont +"', Small='"+ Small +"', RSVAC='"+ RSVAC +"', RSVNAC='"+ RSVNAC +"', Medium='"+ Medium +"', RMVAC='"+ RMVAC +"', RMVNAC='"+ RMVNAC +"', Large='"+ Large +"', RLVAC='"+ RLVAC +"',  RLVNAC='"+ RLVNAC +"', Support='"+ Support +"', RSAC='"+ RSAC +"', RSNAC='"+ RSNAC +"' where id='"+ id +"'")
        cursor.execute("commit");
        show() #calling show function
        MessageBox.showinfo("Update Status", "Updated successfully");
        e_id.delete(0, 'end')
        e_name.delete(0, 'end')
        e_loc.delete(0, 'end')
        e_cont.delete(0, 'end')
        e_small.delete(0, 'end')
        e_rsvac.delete(0, 'end')
        e_rsvnac.delete(0, 'end')
        e_mid.delete(0, 'end')
        e_rmvac.delete(0, 'end')
        e_rmvnac.delete(0, 'end')
        e_large.delete(0, 'end')
        e_rlvac.delete(0, 'end')
        e_rlvnac.delete(0, 'end')
        e_support.delete(0, 'end')
        e_rsac.delete(0, 'end')
        e_rsnac.delete(0, 'end')
        con.close();
            
    def show(): #funct for showing the hid and name on the manupulating hsopital data screen
        con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
        cursor = con.cursor()
        cursor.execute("select * from ambulance")
        data = cursor.fetchall()
        list.delete(0, list.size())
        
        for row in data:
            insertData = '     ' + row[0]+ '             '+ row[1]
            list.insert(list.size()+1, insertData)
        con.close()
        
        
        
        
        
        
    ambu1 = Tk()
    ambu1.title("Manipulate Ambulance Data")
    ambu1.geometry('700x700+450+70')
    h3=tkinter.Label(ambu1,text="Manipulate Ambulance Data",font=("Times new roman",20),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h3.pack(side=TOP,fill=X)
    
    id = Label(ambu1,text="Enter id",font=('bold',10))
    id.place(x=20,y=45)
    e_id = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_id.place(x=180,y=45)
    name = Label(ambu1,text="Enter Name",font=('bold',10))
    name.place(x=20,y=75)
    e_name = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_name.place(x=180,y=75)
    loc = Label(ambu1,text="Enter Location",font=('bold',10))
    loc.place(x=20,y=105)
    e_loc = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_loc.place(x=180,y=105)
    cont = Label(ambu1,text="Enter Contact",font=('bold',10))
    cont.place(x=20,y=135)
    e_cont = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_cont.place(x=180,y=135)
    small = Label(ambu1,text="Available Small Vehicles",font=('bold',10))
    small.place(x=20,y=165)
    e_small = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_small.place(x=180,y=165)
    rsvac = Label(ambu1,text="Rate_SV_AC",font=('bold',10))
    rsvac.place(x=20,y=195)
    e_rsvac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rsvac.place(x=180,y=195)
    rsvnac = Label(ambu1,text="Rate_SV_NonAC",font=('bold',10))
    rsvnac.place(x=20,y=225)
    e_rsvnac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rsvnac.place(x=180,y=225)
    mid = Label(ambu1,text="Available Mid Vehicles",font=('bold',10))
    mid.place(x=20,y=255)
    e_mid = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_mid.place(x=180,y=255)
    rmvac = Label(ambu1,text="Rate_MV_AC",font=('bold',10))
    rmvac.place(x=20,y=285)
    e_rmvac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rmvac.place(x=180,y=285)
    rmvnac = Label(ambu1,text="Rate_MV_NonAC",font=('bold',10))
    rmvnac.place(x=20,y=315)
    e_rmvnac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rmvnac.place(x=180,y=315)
    large = Label(ambu1,text="Available Large Vehicles",font=('bold',10))
    large.place(x=20,y=345)
    e_large = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_large.place(x=180,y=345)
    rlvac = Label(ambu1,text="Rate_LV_AC",font=('bold',10))
    rlvac.place(x=20,y=375)
    e_rlvac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rlvac.place(x=180,y=375)
    rlvnac = Label(ambu1,text="Rate_LV_NonAC",font=('bold',10))
    rlvnac.place(x=20,y=405)
    e_rlvnac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rlvnac.place(x=180,y=405)
    support = Label(ambu1,text="Available Support Vehicles",font=('bold',10))
    support.place(x=20,y=435)
    e_support = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_support.place(x=180,y=435)
    rsac = Label(ambu1,text="Rate_S_AC",font=('bold',10))
    rsac.place(x=20,y=465)
    e_rsac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rsac.place(x=180,y=465)
    rsnac = Label(ambu1,text="Rate_S_NonAC",font=('bold',10))
    rsnac.place(x=20,y=495)
    e_rsnac = Entry(ambu1,font=('ariel' ,10),bd=2,relief=RIDGE)
    e_rsnac.place(x=180,y=495)
    
    btninsert=Button(ambu1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Insert", bg="azure3", command=insert)
    btninsert.place(x=30, y=600)
    btnupdate=Button(ambu1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Update", bg="azure3", command=update)
    btnupdate.place(x=140, y=600)
    btndelete=Button(ambu1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Delete", bg="azure3", command=delete)
    btndelete.place(x=250, y=600)
    btnget=Button(ambu1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Get", bg="azure3", command=get)
    btnget.place(x=360, y=600)
    btnexit=Button(ambu1,padx=7,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Exit", bg="azure3", command=ambu1.destroy)
    btnexit.place(x=470, y=600)
    
    list = Listbox(ambu1,width=35, height=15)
    list.place(x=450,y=45)
    show()
    
    ambu1.mainloop()
    
    
def admin():
    #root.destroy()
    def ok():
        login = e_login.get();
        password = e_pass.get();
        
        if(login=='' or password==''):
            MessageBox.showinfo("Alert", "All fields are required")
        elif(login=='admin' and password=='123'):
            MessageBox.showinfo("Great", "Login Successful")
            admin1.destroy()
            def login():
                log = Tk()
                log.geometry("800x500+350+145")
                log.title("Admin")
                log.configure(background='azure3')
                h2=tkinter.Label(log,text="Welcome As Admin",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
                h2.pack(side=TOP,fill=X)
                
                btnindi=Button(log,padx=50,pady=10,fg="black",font=('ariel' ,10,'bold'),width=8, text="Individual Inserted Data", bg="azure3", command=showind)
                btnindi.place(x=40, y=100)
                
                btnhosp=Button(log,padx=50,pady=10,fg="black",font=('ariel' ,10,'bold'),width=8, text="Manipulate Hospital Data", bg="azure3", command=hosp)
                btnhosp.place(x=500, y=100)
                
                btnambu=Button(log,padx=65,pady=10,fg="black",font=('ariel' ,10,'bold'),width=8, text="Manipulate Ambulance Data", bg="azure3",command=ambu)
                btnambu.place(x=250, y=220)
                
                log.mainloop()
            login()    #Function Call
        else:
            MessageBox.showinfo("Alert", "Incorrect Loginid or Password")
    
    admin1 = tkinter.Tk()
    admin1.geometry('500x350+500+227')
    admin1.configure(background='azure2')
    
    h1=tkinter.Label(admin1,text="ADMIN LOGIN",font=("Times new roman",20),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h1.pack(side=TOP,fill=X)
    
    login=Label(admin1, font=( 'azure2' ,10, 'bold' ),text="Login id",bd=8)
    login.place(x=20,y=110)
    e_login= Entry(admin1,font=('azure2' ,10),bd=4,relief=RIDGE)
    e_login.place(x=128,y=110)
    
    password=Label(admin1, font=( 'azure2' ,10, 'bold' ),text="Password",bd=8)
    password.place(x=20,y=155)
    e_pass= Entry(admin1,font=('azure2' ,10),bd=4,relief=RIDGE)
    e_pass.place(x=128,y=155)
    e_pass.config(show='*')
    
    btnlogin=Button(admin1,padx=20,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Login", bg="azure3", command=ok)
    btnlogin.place(x=40, y=250)
    
    btnexit=Button(admin1,padx=20,pady=1,fg="black",font=('ariel' ,10,'bold'),width=8, text="Exit", bg="azure3", command=admin1.destroy)
    btnexit.place(x=180, y=250)
    
    admin1.mainloop()


root=tkinter.Tk()
root.geometry("800x500+350+145")
root.title("COVID-19 RELIEF")
root.configure(background='azure3')

h=tkinter.Label(root,text="COVID-19 RELIEF CAMP",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
h.pack(side=TOP,fill=X)

Tops = Frame(root,bg="white",width = 1600,height=50,relief=SUNKEN)
Tops.pack(side=TOP)

#------------------TIME--------------
localtime=time.asctime(time.localtime(time.time()))
#-----------------INFO TOP------------
lblinfo = Label(Tops, font=( 'aria' ,20, ),text=localtime,fg="gray7",anchor=W)
lblinfo.grid(row=1,column=0)


def bg():
    def search():
        add = e_add.get()
        bg = e_bg.get()
        if(add=='' or bg==''):
            MessageBox.showinfo("Alert", "All fields are required")
        else:
            search1 = Tk()
            search1.title("BG Details")
            search1.geometry('1000x400+200+100')

            p1=Label(search1, text="Name", font="time 13 bold")
            p1.grid(row=1,column=0,padx=5,pady=5)
            p2=Label(search1, text="Location", font="time 13 bold")
            p2.grid(row=1,column=1,padx=5,pady=5)
            p3=Label(search1, text="Contact", font="time 15 bold")
            p3.grid(row=1,column=2,padx=5,pady=5)
            p4=Label(search1, text="Blood Group", font="time 13 bold")
            p4.grid(row=1,column=3,padx=5,pady=5)
            p5=Label(search1, text="Availablity", font="time 13 bold")
            p5.grid(row=1,column=4,padx=5,pady=5)

            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cur = con.cursor()
            if(bg == " A+"):
                query = " select name,loc,cont, IF(bg1>0,'Available','NA') from hospital;"
                cur.execute(query)
                print(bg)
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' A-'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg2>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' B+'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg3>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' B-'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg4>0,'Available', 'NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' O+'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg8>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' O-'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg7>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            elif(bg==' AB+'):
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg5>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="steel blue", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)

            else:
                print(bg)
                cur.execute(" select name,loc,cont, IF(bg6>0,'Available','NA') from hospital;")
                data = cur.fetchall()
                num=4   
                for i in data:
                    a1=Label(search1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(search1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=1,padx=5,pady=5)
                    a3=Label(search1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=2,padx=5,pady=5)
                    a4=Label(search1, text=bg, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=3,padx=5,pady=5)
                    a5=Label(search1, text=i[3], fg="black", anchor=W)
                    a5.grid(row=num,column=4,padx=5,pady=5)
                    if i[3] =='Available':
                        btnfind=Button(search1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find",command=partial(dijkstra_DSH,graph,str(add),str(i[0])))
                        btnfind.grid(row=num,column=5)
                    num=num+2
                print(data)
            con.close();
    
    #category.destroy()
    bg1=tkinter.Tk()
    bg1.geometry("800x550+350+100")
    bg1.title("COVID-19 RELIEF")
    h=tkinter.Label(bg1,text="Blood Group Category",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h.pack(side=TOP,fill=X)
    Bg=Label(bg1,text="Blood Group Needed:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=20,y=100)
    e_bg = ttk.Combobox(bg1, width=27, state='readonly')
    # Adding combobox drop down list
    e_bg['values'] = (' A+',' A-',' B+',' B-',' O+',' O-',' AB+',' AB-')
    e_bg.place(x= 240, y=114)
    Add=Label(bg1, font=( 'aria' ,14, 'bold' ),text="Address:",bd=10)
    Add.place(x=20,y=180)
    e_add = ttk.Combobox(bg1,width=35,height=10, state='readonly')
    e_add['values'] = ('Jubilee More','ESI Hospitals','HLG','Shristi Nagar','Kalyanpur Housing','Dhadka','Railpaar','Asansol Railway Station','City Bus Stand','Chelidanga','BNR More','Bhagat Singh More','Kumarpur','Neamatpur','Kulti','Barakar','New Town',
                       'Radha Nagar','Galaxy Mall','Burnpur Railway Station','Burnpur Bus Stand','Nehru Park','Sterling Hospital','Court Area',
                       'S.B. Gorai Road','Ismile More','Rehmat Nagar','Hutton Road','Asansol District Hospital','Mohishila Colony','Murgasol',
                       'Ashram More','Kalla More','Ushagram','Kalipahari','Raniganj','Anand Lok Hospital','Durgapur City Centre',
                       'Health World Hospital','I Q Hospital','The Mission'
                      )
    e_add.place(x=240,y=194)
    btnsearch=Button(bg1,padx=60,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Search For Availabilty", bg="azure3", command=search)
    btnsearch.place(x=22, y=250)
    btnexit=Button(bg1,padx=10,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Go Back", bg="azure3", command=bg1.destroy)
    btnexit.place(x=350, y=250)
    bg1.mainloop()

    
def amb(): 
    amb1=Tk()
    amb1.geometry("950x750+280+100")
    amb1.title("Search For Ambulance")
    h=tkinter.Label(amb1,text="Ambulance Services",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h.pack(side=TOP,fill=X)
    source=Label(amb1,text="Select Pick Up:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=20,y=100)
    e_source = ttk.Combobox(amb1, width=27, state='readonly')
    e_source['values'] = ('Jubilee More','ESI Hospitals','HLG','Shristi Nagar','Kalyanpur Housing','Dhadka','Railpaar','Asansol Railway Station','City Bus Stand','Chelidanga','BNR More','Bhagat Singh More','Kumarpur','Neamatpur','Kulti','Barakar','New Town',
                       'Radha Nagar','Galaxy Mall','Burnpur Railway Station','Burnpur Bus Stand','Nehru Park','Sterling Hospital','Court Area',
                       'S.B. Gorai Road','Ismile More','Rehmat Nagar','Hutton Road','Asansol District Hospital','Mohishila Colony','Murgasol',
                       'Ashram More','Kalla More','Ushagram','Kalipahari','Raniganj','Anand Lok Hospital','Durgapur City Centre',
                       'Health World Hospital','I Q Hospital','The Mission'
                      )
    e_source.place(x=210,y=114)
    destination=Label(amb1,text="Select Drop Off:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=500,y=100)
    e_destination = ttk.Combobox(amb1, width=27, state='readonly')
    e_destination['values'] = ('Jubilee More','ESI Hospitals','HLG','Shristi Nagar','Kalyanpur Housing','Dhadka','Railpaar','Asansol Railway Station','City Bus Stand','Chelidanga','BNR More','Bhagat Singh More','Kumarpur','Neamatpur','Kulti','Barakar','New Town',
                       'Radha Nagar','Galaxy Mall','Burnpur Railway Station','Burnpur Bus Stand','Nehru Park','Sterling Hospital','Court Area',
                       'S.B. Gorai Road','Ismile More','Rehmat Nagar','Hutton Road','Asansol District Hospital','Mohishila Colony','Murgasol',
                       'Ashram More','Kalla More','Ushagram','Kalipahari','Raniganj','Anand Lok Hospital','Durgapur City Centre',
                       'Health World Hospital','I Q Hospital','The Mission'
                      )
    e_destination.place(x=690,y=114)
    
    def check():
        def Dijkstra_Ambu(graph,start,goal):
            shortest_distance = {} #records the cost to reach to that node...Going to be updated as we move along the graph
            track_predecessor = {} #keep track of the path that has led us to this
            unseenNodes = graph #to iterate through the entire graph
            infinity = 99999 #it can be considered as a very large number
            track_path = [] #going to trace our journey to the source node...optimal route

            for node in unseenNodes:
                shortest_distance[node] = infinity
            shortest_distance[start] = 0

            while unseenNodes:
                min_distance_node = None
                for node in unseenNodes:
                    if min_distance_node is None:
                        min_distance_node = node
                    elif shortest_distance[node] < shortest_distance[min_distance_node]:
                        min_distance_node = node

                path_options = graph[min_distance_node].items()

                for child_node, weight in path_options:
                    if weight + shortest_distance[min_distance_node] < shortest_distance[child_node]:
                        shortest_distance[child_node] = weight + shortest_distance[min_distance_node]
                        track_predecessor[child_node] = min_distance_node

                unseenNodes.pop(min_distance_node)

            currentNode = goal

            while currentNode != start:
                    track_path.insert(0, currentNode)
                    currentNode = track_predecessor[currentNode]

            track_path.insert(0, start)
            km1 = shortest_distance[goal]
            print("KM1="+str(km1))
            return km1
        
        def Findambu(km):
            
            def costcalc(rate,km): 
                num=0
                F1 = Tk()
                F1.title("Ambulance Details")
                F1.geometry('680x300+750+200')
                cost=rate*km
                print("The total cost is "+str(cost)+"@"+str(rate)+"Per KM")
                l1=Label(F1, text="Distance is :", font="times 15 bold")
                l1.grid(row=num+1,column=1,padx=5,pady=5)
                a1=Label(F1, text=km, fg="steel blue", font="times 15 bold", anchor=W)
                a1.grid(row=num+1,column=2,padx=5,pady=5)
                l2=Label(F1, text="Cost is :", font="times 15 bold")
                l2.grid(row=num+3,column=1,padx=5,pady=5)
                a2=Label(F1, text=cost, fg="steel blue", font="times 15 bold", anchor=W)
                a2.grid(row=num+3,column=2,padx=5,pady=5) 
                l3=Label(F1, text="@ Rs.", font="times 15 bold")
                l3.grid(row=num+3,column=3,padx=5,pady=5)
                a3=Label(F1, text=rate, fg="steel blue", font="times 15 bold", anchor=W)
                a3.grid(row=num+3,column=4,padx=5,pady=5)
                l4=Label(F1, text="Per KM", fg="steel blue", font="times 15 bold", anchor=W)
                l4.grid(row=num+3,column=5,padx=5,pady=5)
                if(cat=='Small (Budget Friendly)'):
                    l5=Label(F1, text="Vehicle Provided will be Maruti Suzuki Omnie", font="times 16 bold", fg="royal blue")
                    l5.grid(row=num+6,column=1,padx=5,pady=5)
                elif(cat=='Medium'):
                    l5=Label(F1, text="Vehicle Provided will be Mahindra Bolero", font="times 16 bold", fg="royal blue")
                    l5.grid(row=num+6,column=1,padx=5,pady=5)
                elif(cat=='Large (Spacious)'):
                    l5=Label(F1, text="Vehicle Provided will be Tata Winger", font="times 16 bold", fg="royal blue")
                    l5.grid(row=num+6,column=1,padx=5,pady=5)
                else:
                    l5=Label(F1, text="Vehicle Provided will be Tata Winger", font="times 16 bold", fg="royal blue")
                    l5.grid(row=num+6,column=1,padx=5,pady=5)


            
            
            cat = e_type.get()
            op = e_option.get()
            
            F1 = Tk()
            F1.title("Ambulance Details")
            F1.geometry('600x400+100+150')
            p1=Label(F1, text="Name", font="time 15 bold")
            p1.grid(row=1,column=0,padx=5,pady=5)
            p2=Label(F1, text="Location", font="time 15 bold")
            p2.grid(row=1,column=4,padx=5,pady=5)
            p3=Label(F1, text="Contact", font="time 15 bold")
            p3.grid(row=1,column=8,padx=5,pady=5)
            p4=Label(F1, text="Category", font="time 15 bold")
            p4.grid(row=1,column=12,padx=5,pady=5)
            p5=Label(F1, text="Type", font="time 15 bold")
            p5.grid(row=1,column=16,padx=5,pady=5)
            
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cur = con.cursor()
            if(cat=='Small (Budget Friendly)' and op=='AC'):
                cur.execute("select Name,Loc,Cont,RSVAC, IF(Small>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            #print(data) 
            
            elif(cat=='Small (Budget Friendly)' and op=='Non-AC'):
                cur.execute("select Name,Loc,Cont,RSVNAC, IF(Small>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            #print(data)
            
            elif(cat=='Medium' and op=='AC'):
                cur.execute("select Name,Loc,Cont,RMVAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            elif(cat=='Medium' and op=='Non-AC'):
                cur.execute("select Name,Loc,Cont,RMVNAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
                
            elif(cat=='Large (Spacious)' and op=='AC'):
                cur.execute("select Name,Loc,Cont,RLVAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            
            elif(cat=='Large (Spacious)' and op=='Non-AC'): 
                cur.execute("select Name,Loc,Cont,RLVNAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            
            elif(cat=='Support (Spacious with Attendant)' and op=='AC'):
                cur.execute("select Name,Loc,Cont,RSAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            elif(cat=='Support (Spacious with Attendant)' and op=='Non-AC'):
                cur.execute("select Name,Loc,Cont,RSNAC, IF(Medium>0,'Available','NA') from ambulance;")
                data = cur.fetchall()
                num=4
                for i in data:
                    a1=Label(F1, text=i[0], fg="steel blue", anchor=W)
                    a1.grid(row=num,column=0,padx=5,pady=5)
                    a2=Label(F1, text=i[1], fg="steel blue", anchor=W)
                    a2.grid(row=num,column=4,padx=5,pady=5)
                    a3=Label(F1, text=i[2], fg="steel blue", anchor=W)
                    a3.grid(row=num,column=8,padx=5,pady=5)
                    a4=Label(F1, text=cat, fg="steel blue", anchor=W)
                    a4.grid(row=num,column=12,padx=5,pady=5)
                    a5=Label(F1, text=op, fg="steel blue", anchor=W)
                    a5.grid(row=num,column=16,padx=5,pady=5)
                    if i[4] =='Available':
                        btnfind=Button(F1,padx=10,pady=0.1,fg="black",font=('ariel' ,8,'bold'), text="Find", command=partial(costcalc,i[3],km))
                        btnfind.grid(row=num,column=20)
                    num=num+2
            else:
                MessageBox.showinfo("Alert", "Wrong Input")
                    
                
            con.close();      
            F1.mainloop()
                
        
        s = e_source.get()
        d = e_destination.get()
        if(s == d):
            MessageBox.showinfo("Alert", "Source and Destination can't be same")
        else:
            l=Label(amb1, text="Valid", font="times 25 bold", fg="green3")
            l.place(x=320,y=200)
            
            km=Dijkstra_Ambu(graph,s,d)
            time.sleep(1)
            
            type = Label(amb1,text="Select Type:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=20,y=300)
            e_type = ttk.Combobox(amb1, width=30, state='readonly')
            e_type['values'] = ('Small (Budget Friendly)','Medium','Large (Spacious)','Support (Spacious with Attendant)')
            e_type.place(x=210,y=314)
            option=Label(amb1,text="Select Option:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=500,y=300)
            e_option = ttk.Combobox(amb1, width=27, state='readonly')
            e_option['values'] = ('AC','Non-AC')
            e_option.place(x=690,y=314)
            btnfind=Button(amb1,padx=10,pady=1,fg="black",font=('ariel' ,16,'bold'),width=5, text="Find", bg="azure3",command=partial(Findambu,km))
            btnfind.place(x=30, y=410)
            btnexit=Button(amb1,padx=10,pady=1,fg="black",font=('ariel' ,16,'bold'),width=5, text="Go Back", bg="azure3", command=amb1.destroy)
            btnexit.place(x=240, y=410)
            
            
            
    
    btncheck=Button(amb1,padx=30,pady=1,fg="black",font=('ariel' ,16),width=10, text="Check For Validity", bg="azure3",command=check)
    btncheck.place(x=30, y=200)
    amb1.mainloop()



def Ind(): #Registration Form page
    
    def submit():  #For submitting the data of the form
        Name = e_name.get()
        Age = e_age.get()
        Add = e_add.get()
        Gen = e_gen.get()
        Cont = e_cont.get()
        Email = e_email.get()
        Date = e_date.get()
    
    
        if (Name=='' or Age=='' or Add=='' or Gen=='' or Cont=='' or Email=='' or Date==''):
            MessageBox.showinfo("Insert Status", "All fields are required")
        else:
            con = mysql.connect(host="localhost", user="root", password="", database="trial 2")
            cursor = con.cursor()
            cursor.execute("insert into form values('"+ Name +"','"+ Age +"','"+ Add +"','"+ Gen +"','"+ Cont +"','"+ Email +"','"+ Date +"')")
            cursor.execute("commit");
            MessageBox.showinfo("Insert Status", "Inserted successfully");
            con.close();
            
        
        #After Final coding, Put these below 13 lines under the else section
        category=tkinter.Tk()
        category.geometry("1050x550+220+100")
        category.title("COVID-19 RELIEF")
        h=tkinter.Label(category,text="Choose The Category",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
        h.pack(side=TOP,fill=X)
        btnbg=Button(category,padx=40,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Need Blood Group", bg="azure3", command=bg)
        btnbg.place(x=80, y=180)
        btnamb=Button(category,padx=40,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Ambulance Service", bg="azure3", command=amb)
        btnamb.place(x=400, y=180)
        btnmed=Button(category,padx=40,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Medicine Support", bg="azure3")
        btnmed.place(x=720, y=180)
        btnexit=Button(category,padx=10,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Exit", bg="azure3", command=category.destroy)
        btnexit.place(x=420, y=400)
        
        
        
    
    def clear():
        e_name.delete(0, 'end')
        e_age.delete(0, 'end')
        e_add.set('')
        #e_bg.set('')
        e_gen.set('')
        e_cont.delete(0, 'end')
        e_email.delete(0, 'end')
        e_date.delete(0, 'end')
    
    
        
        
    
    window=tkinter.Tk()
    window.geometry("600x650+450+70")
    window.title("COVID-19 RELIEF")
    h=tkinter.Label(window,text="Registration Form",font=("Times new roman",30),fg='black',bg='RoyalBlue4',relief=tkinter.GROOVE)
    h.pack(side=TOP,fill=X)
    
    Name=Label(window, font=( 'aria' ,16, 'bold' ),text="Name:",bd=10)
    Name.place(x=20,y=65)
    e_name= Entry(window,font=('ariel' ,14),bd=4,relief=RIDGE)
    e_name.place(x=128,y=75)
    
    Age=Label(window, font=( 'aria' ,14, 'bold' ),text="Age:",bd=10)
    Age.place(x=20,y=125)
    e_age= Entry(window,font=('ariel' ,14),bd=4,relief=RIDGE)
    e_age.place(x=128,y=130)
    
    Add=Label(window, font=( 'aria' ,14, 'bold' ),text="Address:",bd=10)
    Add.place(x=20,y=185)
    e_add = ttk.Combobox(window,width=35,height=10, state='readonly')
    e_add['values'] = ('Jubilee More','ESI Hospitals','HLG','Shristi Nagar','Kalyanpur Housing','Dhadka','Railpaar','Asansol Railway Station','City Bus Stand','Chelidanga','BNR More','Bhagat Singh More','Kumarpur','Neamatpur','Kulti','Barakar','New Town',
                       'Radha Nagar','Galaxy Mall','Burnpur Railway Station','Burnpur Bus Stand','Nehru Park','Sterling Hospital','Court Area',
                       'S.B. Gorai Road','Ismile More','Rehmat Nagar','Hutton Road','Asansol District Hospital','Mohishila Colony','Murgasol',
                       'Ashram More','Kalla More','Ushagram','Kalipahari','Raniganj','Anand Lok Hospital','Durgapur City Centre',
                       'Health World Hospital','I Q Hospital','The Mission'
                      )
    e_add.place(x=128,y=190)
    
    #Bg=Label(window,text="Blood Group Needed:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=20,y=200)
    #e_bg = ttk.Combobox(window, width=27, state='readonly')
    # Adding combobox drop down list
    #e_bg['values'] = (' A+',' A-',' B+',' B-',' O+',' O-',' AB+',' AB-')
    #e_bg.place(x= 240, y=212)

    Gen=Label(window,text="Gender:", font=("aerial", 14, "bold"), bd=10, justify='right').place(x=20,y=230)
    e_gen = ttk.Combobox(window, width=27, state='readonly')
    # Adding combobox drop down list
    e_gen['values'] = (' Male',' Female',' Others')
    e_gen.place(x= 128, y=240)

    Cont=Label(window, font=( 'aria' ,14, 'bold' ),text="Contact:",bd=10)
    Cont.place(x=20,y=278)
    e_cont= Entry(window, font=('ariel' ,14))
    e_cont.place(x=128,y=288)

    Email=Label(window, font=( 'aria' ,14, 'bold' ),text="Email:",bd=10)
    Email.place(x=20,y=324)
    e_email= Entry(window, font=('ariel' ,14))
    e_email.place(x=128,y=334)
    
    Date=Label(window, font=( 'aria' ,14, 'bold' ),text="Date:",bd=10)
    Date.place(x=20,y=370)
    e_date= Entry(window,font=('ariel' ,14))
    e_date.place(x=128,y=380)
    
    
    btnsubmit=Button(window,padx=20,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Submit", bg="azure3", command=submit)
    btnsubmit.place(x=190, y=460)
    
    BtnExit=Button(window,text='Exit',font=('ariel',15,'bold'),padx=20,pady=1,  bg="azure3", command=window.destroy)
    BtnExit.place(x=380,y=540)
    
    btnclear=Button(window,padx=20,pady=1,fg="black",font=('ariel' ,16,'bold'), text="Clear", bg="azure3", command=clear)
    btnclear.place(x=80, y=540)



    window.mainloop()
    
    

btnind=Button(root,padx=45,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Registration Form", bg="azure3", command=Ind)
btnind.place(x=260, y=180)


btnadmin=Button(root,padx=20,pady=1,fg="black",font=('ariel' ,16,'bold'),width=10, text="Admin Login", bg="azure3", command=admin)
btnadmin.place(x=450, y=300)

BtnExit=Button(root,text='Exit',font=('ariel',15,'bold'),width=10,  bg="azure3", command=root.destroy)
BtnExit.place(x=150,y=300)


root.mainloop()
 
```
